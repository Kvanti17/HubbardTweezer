## Test where minimization converges when different weights are assigned

````python
from Hubbard.plot import *
import numpy as np
import matplotlib.pyplot as plt


N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([4], dtype=int),
                 band=1,
                 dim=1,
                 avg=1 / 2,
                 sparse=True,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)
````

Varying $U$ weight

````python
G.eq_label = 'eq'
a = np.arange(0, 1.1, 0.1)
a = np.append(a, [2, 5, 10])
b = np.array([])
c = np.array([])
for alpha in a:
    __, __, info = G.equalzie(
        'uVt', weight=np.array([alpha, 1, 1]), callback=True)
    b = np.append(b, info['ctot'][-1])
    c = np.append(c, info['fval'][-2])

plt.plot(a, b)
plt.plot(a, c)
plt.xlabel('U weight')
plt.ylabel('Cost function value')
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(['total cost', 'min target'])
````

`total cost` means the cost function of $c_U+c_t+c_V$, `min target` is the cost function used in minimization $\alpha c_U + \beta c_t + \gamma c_V$, with the weights adjusted.

Left end of the plot:

````python
__, __, info = G.equalzie('Vt', callback=True)
G.singleband_Hubbard(u=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

Right end:

````python
__, __, info = G.equalzie('uVt', weight=np.array([10, 1, 1]), callback=True)
G.singleband_Hubbard(u=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

Varying $t$ weight

````python
G.eq_label = 'eq'
a = np.arange(0, 1.1, 0.1)
b = np.array([])
c = np.array([])
for alpha in a:
    __, __, info = G.equalzie(
        'Uvt', weight=np.array([1, alpha, 1]), callback=True)
    b = np.append(b, info['ctot'][-1])
    c = np.append(c, info['fval'][-2])

plt.plot(a, b)
plt.plot(a, c)
plt.xlabel('t weight')
plt.ylabel('Cost function value')
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(['total cost', 'min target'])
````

Varying $V$ weight

````python
G.eq_label = 'eq'
a = np.arange(0, 1.1, 0.1)
a = np.append(a, [2, 5, 10])
b = np.array([])
c = np.array([])
for alpha in a:
    __, __, info = G.equalzie(
        'Uvt', weight=np.array([1, 1, alpha]), callback=True)
    b = np.append(b, info['ctot'][-1])
    c = np.append(c, info['fval'][-2])

plt.plot(a, b)
plt.plot(a, c)
plt.xlabel('V weight')
plt.ylabel('Cost function value')
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(['total cost', 'min target'])
````

Plot left end:

````python
__, __, info = G.equalzie('Ut', callback=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

It seems that $V$ and $U$ are competing. The total cost function minimum is flat. This agrees with the observation in `equalization.ipynb` that when equalizing only $U$, the trap depth fluctuation increases.

Plot right end:

````python
__, __, info = G.equalzie('Uvt', weight=np.array([1, 1, 10]), callback=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

````python
from multiprocessing.dummy import Pool
from Hubbard.plot import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([3], dtype=int),
                 band=1,
                 dim=1,
                 avg=1 / 2,
                 sparse=True,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)

u, t, v = True, True, True
fix_u, fix_v = True, False

res = G.singleband_Hubbard(u=u, output_unitary=True)
if u:
    A, U, V = res
else:
    A, V = res
    U = None

if fix_u:
    Utarget = np.mean(U)
else:
    Utarget = None
if t:
    nnt = G.nn_tunneling(A)
    xlinks, ylinks, nntx, nnty = G.xy_links(nnt)
else:
    nnt, xlinks, ylinks, nntx, nnty = None, None, None, None, None
if fix_v:
    Vtarget = np.mean(np.real(np.diag(A)))
else:
    Vtarget = None

v0, bounds = G.init_guess()
x0 = [V]


def cost_func(offset: np.ndarray, info: Union[dict, None]) -> float:
    c = G.cbd_cost_func(offset, info, (xlinks, ylinks),
                        (Vtarget, Utarget, nntx, nnty), (u, t, v), unitary=x0)

    return c
````

````python
Ntime = 10
x = np.linspace(bounds[0][0], bounds[0][1], Ntime)
y = np.linspace(bounds[2][0], bounds[2][1], Ntime)
Z = np.zeros((len(x), len(y)))
X, Y = np.meshgrid(x, y)
for i in range(len(x)):
    for j in range(len(y)):
        offset = np.array([x[i], 1, y[j], 0, 0, 0])
        Z[i, j] = cost_func(offset, None)
````

````python
import matplotlib.pyplot as plt
import matplotlib.colors as colors

plt.pcolormesh(X, Y, Z, norm=colors.LogNorm(vmin=Z.min(), vmax=Z.max()))
plt.colorbar()
````

## Test where minimization converges when different weights are assigned

````python
from Hubbard.plot import *
import numpy as np
import matplotlib.pyplot as plt


N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([4], dtype=int),
                 band=1,
                 dim=1,
                 avg=1 / 2,
                 sparse=True,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)
````

### Varying $U$ weight

````python
G.eq_label = 'eq'
a = np.arange(0, 1.1, 0.1)
a = np.append(a, [2, 5, 10])
b = np.array([])
c = np.array([])
for alpha in a:
    __, __, info = G.equalzie(
        'uVt', weight=np.array([alpha, 1, 1]), callback=True)
    b = np.append(b, info['ctot'][-1])
    c = np.append(c, info['fval'][-2])

plt.plot(a, b)
plt.plot(a, c)
plt.xlabel('U weight')
plt.ylabel('Cost function value')
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(['total cost', 'min target'])
````

`total cost` means the cost function of $c_U+c_t+c_V$, `min target` is the cost function used in minimization $\alpha c_U + \beta c_t + \gamma c_V$, with the weights adjusted.

Left end of the plot:

````python
__, __, info = G.equalzie('Vt', callback=True)
G.singleband_Hubbard(u=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

Right end:

````python
__, __, info = G.equalzie('uVt', weight=np.array([10, 1, 1]), callback=True)
G.singleband_Hubbard(u=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
cv = G.v_cost_func(G.A, None)
ct = G.t_cost_func(G.A, None, None)
cu = G.u_cost_func(G.U, None)
print(f'V_err = {cv}')
print(f't_err = {ct}')
print(f'U_err = {cu}')
print(f'total_err = {cv + ct + cu}')
````

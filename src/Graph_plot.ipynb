## 2D illustration: un-equalized

```python
from plot_Hubbard import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
G = Graph(N,
          R0=R0,
          lattice=np.array([2, 2], dtype=int),
          band=1,
          dim=3,
          avg=1 / 2,
          sparse=True,
          symmetry=True)

G.singleband_params()
G.add_nnn()
G.update_edge_weight()
G.update_node_weight()

nx.draw_networkx_nodes(G.graph,
                       pos=G.pos,
                       node_color='#99CCFF',
                       node_size=G.node_size)
nx.draw_networkx_labels(G.graph,
                        pos=G.pos,
                        font_color='#000066',
                        font_size=8,
                        labels=G.node_label)
link_list = list(G.graph.edges)
for i in range(len(link_list)):
    e = [link_list[i]]
    nx.draw_networkx_edges(G.graph,
                           G.pos,
                           edgelist=e,
                           edge_color='#606060',
                           alpha=np.sqrt(G.edge_alpha[i]),
                           width=3)
nx.draw_networkx_edge_labels(G.graph,
                             G.pos,
                             font_size=10,
                             edge_labels=G.edge_label,
                             font_color=[0.256, 0.439, 0.588])
# alphas = [0.3, 0.4, 0.5]
# for i, arc in enumerate(arcs):  # change alpha values of arcs
#     arc.set_alpha(alphas[i])
```

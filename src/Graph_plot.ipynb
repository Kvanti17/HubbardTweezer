```python
from plot_Hubbard import *
import numpy as np
from positify import positify

N = 20
R0 = np.array([3, 3, 7.2])
G = Graph(N,
          R0=R0,
          lattice=np.array([4], dtype=int),
          lc=(1350, 1500),
          trap=(50, [930, 1250]),
          band=1,
          dim=3,
          avg=1,
          sparse=True,
          symmetry=True)

import matplotlib.pyplot as plt
import networkx as nx

G.Voff = np.array([1.0192159, 0.98180606, 0.98180606, 1.0192159])
G.trap_centers = np.array([[-1.5, 0.], [-0.5, 0.], [0.5, 0.], [1.5, 0.]])
G.update_edge_len()
G.update_node_weight()

pos = dict((n, G.trap_centers[n]) for n in G.graph.nodes())
edge_label = dict(
    (edge, f'{G.graph[edge[0]][edge[1]]["weight"]:.3g}') for edge in G.edges)
node_label = dict((n, f'{G.Voff[n]:.3g}') for n in G.graph.nodes())
nx.draw_networkx(G.graph,
                 pos=pos,
                 edge_color='#606060',
                 width=3,
                 with_labels=False,
                 node_size=[i**10 * 600 for i in G.Voff],
                 node_color='#99CCFF')
pos_labels = G.pos
# pos_labels = {}
# offset = 0.01
# for k, v in pos.items():
#     pos_labels[k] = (v[0], v[1] + offset)
nx.draw_networkx_labels(G.graph,
                        pos=pos_labels,
                        font_size=8,
                        font_color='#00994C',
                        labels=G.node_label)
nx.draw_networkx_edge_labels(G.graph,
                             pos,
                             edge_labels=edge_label,
                             font_color=[0.256, 0.439, 0.588])
# plt.axis('off')
# plt.savefig('Graph.pdf')
```

### Two-band 3x3 square with balancing

```python
from HubbardTweezer.Hubbard.plot import *
import numpy as np
from HubbardTweezer.Hubbard.io import *

N = 20
L = 3
Ly = 3
d = 1
shape = "square"
w = None
eqt = "neq"
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(
    N,
    R0=R0,
    lattice=np.array([L, Ly], dtype=int),
    trap=(52.26, 1000),
    lc=(1550, 1600),
    ascatt=1770,
    laser=800,
    band=2,
    sparse=True,
    shape=shape,
    waist=w,
    equalize=False,
    eqtarget=eqt,
    symmetry=True,
    verbosity=3,
)
fn = "../../output/Hubbard/3D_3x3_square_2bands.ini"
b = 1
read_file(fn, G, band=b)
G.draw_graph(band=b, A=G.A, U=G.U)
```

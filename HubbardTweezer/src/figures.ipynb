### Two-band 3x3 square

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.io import *

N = 20
L = 3
Ly = 3
d = 1
shape = 'square'
w = None
eqt = 'neq'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 trap=(52.26, 1000),
                 lc=(1550, 1600),
                 ascatt=1770,
                 laser=780,
                 band=2,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 eqtarget=eqt,
                 symmetry=True,
                 verbosity=3)
eig_sol = G.eigen_basis()
```

```python
A, W = multiband_WF(G, *eig_sol)
U = interaction(G, W, *eig_sol[1:])
G.draw_graph(band=[1, 2], A=A, U=U)
```

```python
G.singleband_Hubbard(u=True, band=1, eig_sol=eig_sol)
G.draw_graph(band=1, A=G.A, U=G.U)
G.wf_centers

G.singleband_Hubbard(u=True, band=2, offset=G.zero, eig_sol=eig_sol)
G.draw_graph(band=2, A=G.A, U=G.U)
G.wf_centers
```

## 20 site ring

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.io import *

N = 20
L = 20
Ly = 1
d = 3
shape = 'ring'
w = None
eqt = 'neq'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 trap=(52.26, 1000),
                 lc=1550,
                 ascatt=1770,
                 laser=780,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)
fn = f'../output/{d}D_{L}x{Ly}_{shape}_{w}_{eqt}.ini'
report = rep.get_report(fn)
G.U, G.A = read_Hubbard(report)
G.draw_graph(nnn=False, A=G.A, U=G.U)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
print(f'V_err = {G.v_cost_func(G.A, None)}')
print(f't_err = {G.t_cost_func(G.A, None, None, None)}')
print(f'U_err = {G.u_cost_func(G.U, None)}')
```

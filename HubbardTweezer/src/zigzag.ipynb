```python
import matplotlib.pyplot as plt
from Hubbard.plot import *
import numpy as np
from Hubbard.io import *

N = 20
L = 4
Ly = 2
d = 3
shape = 'zigzag'
w = None
eqt = 'vT'
meth = 'trf'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 trap=(73.0219, 1000),
                 lc=(2400, 1000),
                 ascatt=1770,
                 laser=780,
                 band=1,
                 dim=d,
                 avg=1,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 eqtarget=eqt,
                 lattice_symmetry=False,
                 symmetry=True,
                 verbosity=0)
fn = f'../output/{d}D_{L}x{Ly}_{shape}_{w}_{eqt}_{meth}.ini'
report = rep.get_report(fn)
G.U, G.A = read_Hubbard(report)
c = rep.a(report, "Equalization_Log", "cost_func_value")
G.Voff, G.trap_centers, G.waists,  G.sf = read_trap(report)

plt.plot(c, '-')
plt.yscale('log')

G.draw_graph(A=G.A, U=G.U)
print(c[-1])
print(c[-2] - c[-1])
```

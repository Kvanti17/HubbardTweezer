# Lattice geometries

```python
import matplotlib.pyplot as plt
from Hubbard.core import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
dvr = MLWF(N,
           R0=R0,
           lattice=np.array([8], dtype=int),
           lc=2000,
           band=1,
           dim=3,
           avg=1 / 2,
           sparse=True,
           symmetry=True)

x = np.linspace(-dvr.R0[0] * 0.8, dvr.R0[0] * 0.8, 500)
X = np.meshgrid(x, x)
V = dvr.Vfun(*X, 0)

plt.pcolormesh(*X, V)
```

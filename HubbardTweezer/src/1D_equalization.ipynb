### 8x1 chain no waist

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.output import *

N = 20
L = 8
Ly = 1
d = 3
shape = 'square'
w = None
eqt = 'neq'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 band=1,
                 dim=d,
                 avg=0.5,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 eqtarget=eqt,
                 symmetry=True,
                 verbosity=0)
fn = f'../output/{d}D_{L}x{Ly}_{shape}_{w}_{eqt}.ini'
report = rep.get_report(fn)
G.U, G.A = read_Hubbard(report)
G.Voff, G.trap_centers, G.waists, G.sf = read_trap(report)
G.draw_graph(nnn=True, A=G.A, U=G.U)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
nnt = G.nn_tunneling(G.A)
xlinks, ylinks, txTarget, tyTarget = G.xy_links(nnt)
print(f'V_err = {np.sqrt(G.v_cost_func(G.A, None, txTarget))}')
print(
    f't_err = {np.sqrt(G.t_cost_func(G.A, (xlinks, ylinks), (txTarget, tyTarget)))}')
print(f'U_err = {np.sqrt(G.u_cost_func(G.U, None, txTarget))}')
```

```python
point = np.array([0.99107375, 1.00888941, -1.5019168, -0.49426382])
G.sf = .582834667050257
u, t, v = True, True, False
w = np.array([u, t, v])
v0, bounds = G.init_guess(random=False, nobounds=False, lsq=True)
G.eqinfo = {'Nfeval': 0,
            'cost': np.array([]).reshape(0, 3),
            'ctot': np.array([]),
            'fval': np.array([]),
            'diff': np.array([]),
            'x': np.array([]).reshape(0, *v0.shape)}

scale_factor = G.sf
info = G.eqinfo

res = G.singleband_Hubbard(
    u=u, x0=None, offset=True, output_unitary=True)
if u:
    A, U, x0 = res
else:
    A, x0 = res
    U = None
Utarget = np.mean(U)
Vtarget = None
# x0 is used to update unitary[0] in the next iteration

# Print out Hubbard parameters
if G.verbosity > 1:
    print(f'scale_factor = {scale_factor}')
    print(f'V = {np.diag(A)}')
    print(f't = {abs(G.nn_tunneling(A))}')
    print(f'U = {U}')

nnt = G.nn_tunneling(A)
xlinks, ylinks, txTarget, tyTarget = G.xy_links(nnt)
cu = np.zeros(G.Nsite)
if u:
    # U is different, as calculating U costs time
    cu = G.u_res_func(U, Utarget, scale_factor)

cv = G.v_res_func(A, Vtarget, scale_factor)
if not v:
    # Force V to have no effect on cost function
    w[2] = 0

ct = G.t_res_func(A, (xlinks, ylinks), (txTarget, tyTarget))
if not t:
    # Force t to have no effect on cost function
    w[1] = 0

cvec = np.array([la.norm(cu), la.norm(ct), la.norm(cv)])
# Weighted cost function, weight is in front of each squared term
c = np.concatenate(
    [np.sqrt(w[0]) * cu, np.sqrt(w[1]) * ct, np.sqrt(w[2]) * cv])
# The cost func val in least_squares is fval**2 / 2
fval = la.norm(c)
G._update_log(point, info, report, cvec, fval)

G.singleband_Hubbard(u=True)
print(f'V = {np.diag(G.A)}')
cv = G.v_res_func(G.A, None, G.sf)
ct = G.t_res_func(G.A, None, None)
cu = G.u_res_func(G.U, None, G.sf)
print(f't = {abs(G.nn_tunneling(G.A))}')
print(f'U = {G.U}')
print(f'V_err = {la.norm(cv)}')
print(f't_err = {la.norm(ct)}')
print(f'U_err = {la.norm(cu)}')
```

# Numba test

### Code without numba

```python
import numpy as np
from Hubbard.equalizer import *

N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardParamEqualizer(N,
                          R0=R0,
                          lattice=np.array([3], dtype=int),
                          band=1,
                          dim=1,
                          avg=1 / 2,
                          sparse=True,
                          equalize=False,
                          symmetry=True,
                          verbosity=0)
G.eq_label = 'eq'
Ntime = 20
v0, bounds = G.init_guess()
v0_len = v0.shape[0]
c = np.zeros((Ntime, 1 + 2 * v0_len))
for i in range(Ntime):
    __, __, info = G.equalzie('UTv', random=True, callback=True)
    c[i, -1] = info['ctot'][-1]
    c[i, :v0_len] = info['x'][0]
    c[i, v0_len:-1] = info['x'][-1]
```

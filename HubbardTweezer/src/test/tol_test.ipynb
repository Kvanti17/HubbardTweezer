# Algo tolerance test

### Case of `'gtol'`

When cost function has a minimum

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x, a, b):
    return -a * np.exp(-x**2/(2*b)) - a * 0.8 * np.exp(-(x-2)**2/(2*b))

def cost(x, a, b):
    return func(x, a, b) + 1

x = np.linspace(-2, 4, 100)
y = cost(x, 1, 0.5)**2 / 2

plt.plot(x, y, '-', label='data')
```

```python
import numpy as np
from scipy.optimize import minimize, least_squares

least_squares(cost, xtol=None, ftol=None, gtol=1e-6, max_nfev=10, x0=[0], args=(1, 0.5), verbose=2)
```

### Case of `'ftol'`

When cost function has a nonzero minimum and is already at very flat plateau

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x, a, b):
    return -a * np.exp(-x**2/(2*b)) - a * 0.8 * np.exp(-(x-2)**2/(2*b))

def cost(x, a, b):
    return func(x, a, b) + 5

x = np.linspace(-2, 4, 100)
y = cost(x, 1, 0.5)**2 / 2

plt.plot(x, y, '-', label='data')
```

```python
import numpy as np
from scipy.optimize import minimize, least_squares

least_squares(cost, xtol=None, ftol=1e-6, gtol=None, max_nfev=10, x0=[0], args=(1, 0.5), verbose=2)
```

# Case of `'max_iter'`

None of the other criteria is hit

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x, a, b):
    return -a * np.exp(-x**2/(2*b)) - a * 0.8 * np.exp(-(x-2)**2/(2*b))

def cost(x, a, b):
    return func(x, a, b) * np.cos(x * 10) + 1.1

x = np.linspace(-2, 4, 1000)
y = cost(x, 1, 0.5)**2 / 2

plt.plot(x, y, '-', label='data')
```

```python
import numpy as np
from scipy.optimize import minimize, least_squares

least_squares(cost, xtol=None, ftol=1e-12, gtol=1e-8, max_nfev=100, x0=[0], args=(1, 0.5), verbose=2)
```

```python
import numpy as np
from scipy.optimize import minimize, least_squares

least_squares(cost, xtol=None, ftol=1e-12, gtol=1e-8, max_nfev=10, x0=[0], args=(1, 0.5), verbose=2)
```

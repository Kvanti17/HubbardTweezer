Eigenstate check:

Start from the GS of the time-evolving Hamiltonian $H=T+V/2$, then use the same Hamiltonian to do dynamics. Expectedly the state should be kept at GS for the whole time.

```python
from dynamics_sho import *
import matplotlib.pyplot as plt
np.set_printoptions(precision=4, suppress=True)

n = np.array([50, 0, 0])
dx = 1/5 * np.ones(3, dtype=float)
t_list = np.arange(1E-1, 50, 1)
fidelity = []
psi0, E0, W0 = initial_state(n, dx)
for t in t_list:
    U2 = int_evo_ops(n, dx, t, 1/2)
    E2, W2 = sla.schur(U2)
    psi2 = dynamics_by_long_time(t, t, E2, W2, psi0)
    fidelity.append(np.abs(psi2.conj().T @ psi0))
f = plt.figure(figsize=(10, 6))
plt.plot(t_list, fidelity)
plt.title("GS eigen-evolution for $H=T+V/2$")
plt.xlabel('t')
plt.xlabel("t")
plt.ylabel("$|\langle \psi(t) | \psi(0) \\rangle|^2$");
```

The result meets exactly the expectation.

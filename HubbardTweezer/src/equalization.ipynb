## Equalize $U$ only?

### Equalize $U$ by varying depths?

```python
from Hubbard.plot import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([4], dtype=int),
                 band=1,
                 dim=1,
                 avg=1 / 2,
                 sparse=True,
                 equalize=True,
                 eqtarget='u',
                 symmetry=True)
G.draw_graph('adjust')
G.draw_graph()
```

To equalize $U$ bulk/boundary trap depth difference is increased, however to equalzie $V$ it is decreaed to balance with the boundary.

Target should be chosen more wisely, rather than naive mean value.

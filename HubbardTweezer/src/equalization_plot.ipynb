# Equalization illustrations

## 1D 8 site illustrations

```python
from Hubbard.plot import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([8], dtype=int),
                 laser=770,
                 trap=(50, (930, 1250)),
                 lc=(1350, 1550),
                 band=1,
                 dim=1,
                 avg=1,
                 sparse=True,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)
eqtarget = 'uvt'
G.eq_label = eqtarget
__, __, __, info = G.equalzie(eqtarget, callback=True)
G.draw_graph(nnn=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
print(f'V_err = {G.v_cost_func(G.A, None)}')
print(f't_err = {G.t_cost_func(G.A, None, None)}')
print(f'U_err = {G.u_cost_func(G.U, None)}')
```

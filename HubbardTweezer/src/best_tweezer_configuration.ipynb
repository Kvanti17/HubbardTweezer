# Other better trap geometries?

### Idea:
1. Experiments have small trap spacing and large tunnelings
2. But for theory, we know atomic limit is better for tight-binding approx
3. The reason why TB approx fails is that low energy levels are too high for current experiment configurations
4. Possible solution: raise $V_0$ s.t. tunneling is not small, but barrier is higher than energy levels

See notes `best tweezer configuration?`

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.io import *

N = 20
L = 8
Ly = 1
d = 3
shape = 'square'
w = 'x'
eqt = 'UvT'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 trap=(73.0219, 1000),
                 lc=1450,
                 ascatt=1770,
                 laser=780,
                 band=1,
                 dim=d,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=True,
                 eqtarget=eqt,
                 lattice_symmetry=True,
                 symmetry=True,
                 verbosity=0)
G.draw_graph(nnn=True)
```

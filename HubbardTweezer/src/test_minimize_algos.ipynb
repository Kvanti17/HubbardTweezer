# Test least_squares

```python
from Hubbard.equalizer import *
import numpy as np
from Hubbard.output import *

N = 20
L = 4
d = 1
shape = 'square'
w = 'x'
eqt = 'uvt'
R0 = np.array([3, 3, 7.2])
G = HubbardParamEqualizer(N,
                          R0=R0,
                          lattice=np.array([L], dtype=int),
                          band=1,
                          dim=d,
                          avg=0.5,
                          sparse=True,
                          shape=shape,
                          waist=w,
                          equalize=False,
                          symmetry=True,
                          verbosity=0)
G.waist_dir = w
G.equalize_lsq(eqt)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
print(f'V_err = {G.v_cost_func(G.A, None)}')
print(f't_err = {G.t_cost_func(G.A, None, None)}')
print(f'U_err = {G.u_cost_func(G.U, None)}')
```

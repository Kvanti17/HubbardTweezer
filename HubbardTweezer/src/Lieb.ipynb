```python
from Hubbard.plot import *
import numpy as np

N = 20
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([3, 3], dtype=int),
                 band=1,
                 dim=3,
                 avg=0.5,
                 sparse=True,
                 shape='Lieb',
                 waist='xy',
                 equalize=False,
                 symmetry=True,
                 verbosity=2)
eqtarget = 'uvt'
G.eq_label = eqtarget
__, __, __, info = G.equalize(eqtarget, callback=True)
G.draw_graph(nnn=True)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
print(f'V_err = {G.v_cost_func(G.A, None)}')
print(f't_err = {G.t_cost_func(G.A, None, None)}')
print(f'U_err = {G.u_cost_func(G.U, None)}')
```

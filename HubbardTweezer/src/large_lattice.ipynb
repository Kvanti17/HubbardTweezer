# Large lattice Hubbard parameter calculations

## 5x5 Lieb

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.output import *

N = 20
L = 5
d = 3
shape = 'Lieb'
w = None
eqt = 'neq'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, L], dtype=int),
                 band=1,
                 dim=d,
                 avg=0.5,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 symmetry=True,
                 verbosity=0)
fn = f'../output/{d}D_{L}x{L}_{shape}_{w}_{eqt}.ini'
report = rep.get_report(fn)
G.U, G.A = read_parameters(report)
G.draw_graph(nnn=False, A=G.A, U=G.U)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
print(f'V_err = {G.v_cost_func(G.A, None)}')
print(f't_err = {G.t_cost_func(G.A, None, None)}')
print(f'U_err = {G.u_cost_func(G.U, None)}')
```

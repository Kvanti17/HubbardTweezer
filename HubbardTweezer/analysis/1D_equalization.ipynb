### 8x1 chain no waist

```python
from Hubbard.plot import *
import numpy as np
from Hubbard.io import *

N = 20
L = 8
Ly = 1
d = 3
shape = 'square'
w = None
eqt = 'neq'
R0 = np.array([3, 3, 7.2])
G = HubbardGraph(N,
                 R0=R0,
                 lattice=np.array([L, Ly], dtype=int),
                 band=1,
                 dim=d,
                 avg=0.5,
                 sparse=True,
                 shape=shape,
                 waist=w,
                 equalize=False,
                 eqtarget=eqt,
                 symmetry=True,
                 verbosity=0)
fn = f'../output/{d}D_{L}x{Ly}_{shape}_{w}_{eqt}.ini'
report = rep.get_report(fn)
G.U, G.A = read_Hubbard(report)
G.Voff, G.trap_centers, G.waists, G.sf = read_trap(report)
G.draw_graph(nnn=True, A=G.A, U=G.U)
print(f'V = {np.diag(G.A)}')
t = abs(G.nn_tunneling(G.A))
print(f't = {t}')
print(f'U = {G.U}')
nnt = G.nn_tunneling(G.A)
xlinks, ylinks, txTarget, tyTarget = G.xy_links(nnt)
print(f'V_err = {np.sqrt(G.v_cost_func(G.A, None, txTarget))}')
print(
    f't_err = {np.sqrt(G.t_cost_func(G.A, (xlinks, ylinks), (txTarget, tyTarget)))}')
print(f'U_err = {np.sqrt(G.u_cost_func(G.U, None, txTarget))}')
```

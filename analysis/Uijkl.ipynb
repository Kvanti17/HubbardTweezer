```python
from HubbardTweezer.tools import reportIO as rep
import numpy as np

N = 8

fn = f"../../output/Hubbard/Uijkl.ini"
report = rep.get_report(fn)
Uijkl = rep.a(report, "Singleband_Parameters", "U_ijkl")
```

```python
print("Uijkl shape =", Uijkl.shape)
print("Diagonal elements: U_iiii")
for i in range(N):
    print(Uijkl[i, i, i, i])
```

```python
print("Off-diagonal elements: U_ijij")
for i in range(N):
    temp = np.zeros_like(Uijkl[i, i, i, :])
    for j in range(N):
        if j != i:
            temp[j] = Uijkl[i, j, i, j]
    print(temp)
```

```python
print("Off-diagonal elements: U_ijji")
for i in range(N):
    temp = np.zeros_like(Uijkl[i, i, i, :])
    for j in range(N):
        if j != i:
            temp[j] = Uijkl[i, j, j, i]
    print(temp)
```

```python
print("Off-diagonal elements: U_iiij")
for i in range(N):
    temp = np.zeros_like(Uijkl[i, i, i, :])
    for j in range(N):
        if j != i:
            temp[j] = Uijkl[i, i, i, j]
    print(temp)
```

```python
print("Off-diagonal elements: U_iiji")
for i in range(N):
    temp = np.zeros_like(Uijkl[i, i, i, :])
    for j in range(N):
        if j != i:
            temp[j] = Uijkl[i, i, j, i]
    print(temp)
```

```python
print("Off-diagonal elements: U_ijii")
for i in range(N):
    temp = np.zeros_like(Uijkl[i, i, i, :])
    for j in range(N):
        if j != i:
            temp[j] = Uijkl[i, j, i, i]
    print(temp)
```

```python
print("Max off-diagonal elements: U_ijkl")
temp = Uijkl.copy()
for i in range(N):
    temp[i, i, i, i] = 0
maxval, maxidx = np.max(abs(temp)), np.argmax(abs(temp))
print(np.unravel_index(maxidx, temp.shape), maxval)
```
